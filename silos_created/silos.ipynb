```python
import pandas as pd 
import os 
import numpy as np
```

```python
daily_dataset_directory = '../datasets/'
dataset = []
for filename in os.listdir(daily_dataset_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(daily_dataset_directory, filename)
        data = pd.read_csv(file_path)
        dataset.append(data)

combined_data = pd.concat(dataset, ignore_index= True)
       
```
