```python
import pandas as pd 
import os 
import numpy as np
```

```python
RANDOM_SEED = 42
daily_dataset_directory = '../datasets/'
dataset = []
for filename in os.listdir(daily_dataset_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(daily_dataset_directory, filename)
        data = pd.read_csv(file_path)
        dataset.append(data)

centralized_data  = pd.concat(dataset, ignore_index= True)
       
```

### Preprocessing the combined dataset 


```python
print(f"The combined dataset has {centralized_data .shape[0]} number of samples and {centralized_data .shape[1]} number of features")
centralized_data .head(5)
```

```python
centralized_data.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
print(centralized_data.columns)
```

```python
centralized_data  = centralized_data .replace({
    'Web Attack � Brute Force': 'Web Attack Brute Force',
    'Web Attack � XSS': 'Web Attack XSS',
    'Web Attack � Sql Injection': 'Web Attack SQL Injection'
})
```

```python
## Class types in the Label column 
print(centralized_data [' Label'].value_counts())
```

### Checking the number of the null values in the entire dataset 


```python
centralized_data .dropna(subset=['Flow Bytes/s'], inplace = True)
print(centralized_data .isna().sum().to_list())
```

```python
print(f"Centralized Data has {centralized_data.shape[0]} number of samples and {centralized_data.shape[1]} number of features")
```

### Reduced the number of rows by keeping the same class imbalance. 

```python
### Since we are creating the four silos, we will be splitting the entire dataset into 4 different silos with the proper non-IID approach 

non_attack_data = centralized_data[centralized_data[' Label'] == 'BENIGN'].copy()
attack_data = centralized_data[centralized_data[' Label'] != 'BENIGN'].copy()
```

### Creating the silos out of the stratified dataset 

```python
benign_silos = np.array_split(non_attack_data.sample(frac = 0.7, random_state=RANDOM_SEED),4)
```

```python
benign_silosOne = pd.DataFrame(benign_silos[0], columns=centralized_data.columns)
benign_silosTwo = pd.DataFrame(benign_silos[1], columns=centralized_data.columns)
benign_silosThree = pd.DataFrame(benign_silos[2], columns=centralized_data.columns)
benign_silosFour = pd.DataFrame(benign_silos[3], columns=centralized_data.columns)
```

```python
silos_config = {
    1:{
        'attacks': ['DoS Hulk', 'DoS GoldenEye', 'DoS slowloris', 'DoS Slowhttptest'],
        'benign': benign_silosOne
    },
    2:{
        'attacks':['PortScan', 'DDoS', 'Bot'],
        'benign': benign_silosTwo
    },
    3:{
        'attacks':['FTP-Patator', 'SSH-Patator', 'Heartbleed'],
        'benign':benign_silosThree
    },
    4:{
        'attacks':['Web Attack Brute Force', 'Web Attack XSS', 'Web Attack SQL Injection', 'Infiltration'],
        'benign':benign_silosFour
    }
}
```

```python
siloOneConfig = attack_data[attack_data[' Label'].isin(silos_config[1]['attacks'])]
siloOne = pd.concat([siloOneConfig, benign_silosOne], ignore_index=True)
print(siloOne[' Label'].value_counts())
```

```python
siloTwoConfig = attack_data[attack_data[' Label'].isin(silos_config[2]['attacks'])]
siloTwo = pd.concat([siloTwoConfig, benign_silosTwo], ignore_index=True)
print(siloTwo[' Label'].value_counts())
```

```python
siloThreeConfig = attack_data[attack_data[' Label'].isin(silos_config[3]['attacks'])]
siloThree = pd.concat([siloThreeConfig, benign_silosThree], ignore_index=True)
print(siloThree[' Label'].value_counts())
```

```python
siloFourConfig = attack_data[attack_data[' Label'].isin(silos_config[4]['attacks'])]
siloFour = pd.concat([siloFourConfig, benign_silosFour], ignore_index=True)
print(siloFour[' Label'].value_counts())
```

```python
### Dealing with the class imbalance
def class_imbalance(dataframe, label):
    attack_samples = dataframe[dataframe[label] != 'BENIGN']
    benign_samples = dataframe[dataframe[label] == 'BENIGN']
    attack_count = len(attack_samples)

    benign_downsampled = benign_samples.sample(n = attack_count, random_state = RANDOM_SEED)

    balanced = pd.concat([benign_downsampled,attack_samples], ignore_index= True)

    return balanced

```

```python
siloOne = class_imbalance(siloOne,' Label')
print(siloOne[' Label'].value_counts())
siloOne.to_csv('../silos_datasets/siloOne.csv')
```

```python
siloTwo = class_imbalance(siloTwo,' Label')
print(siloTwo[' Label'].value_counts())
siloTwo.to_csv('../silos_datasets/siloTwo.csv')
```
