```python
import pandas as pd 
import os 
import numpy as np
```

```python
daily_dataset_directory = '../datasets/'
dataset = []
for filename in os.listdir(daily_dataset_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(daily_dataset_directory, filename)
        data = pd.read_csv(file_path)
        dataset.append(data)

combined_data = pd.concat(dataset, ignore_index= True)
       
```

### Preprocessing the combined dataset 


```python
print(f"The combined dataset has {combined_data.shape[0]} number of samples and {combined_data.shape[1]} number of features")
combined_data.head(5)
```

```python
combined_data = combined_data.replace({
    'Web Attack � Brute Force': 'Web Attack Brute Force',
    'Web Attack � XSS': 'Web Attack XSS',
    'Web Attack � Sql Injection': 'Web Attack SQL Injection'
})
```
