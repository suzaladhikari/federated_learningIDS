### Creating the silos: We create each silo (client) that has its own local dataset, ideally no direct access to others, and has same feature space(same columns).
### We will be creating different silos (storage tank that keeps things seperated) and then we will be strategiacally partition the dataset across teh simulated clients with non-IID splits.


---
### What is non-IID split?
### Non IID (Independent and Identically Distributed) split is a split in which each client gets different type of data. 

### What does FedAvg do ?
### In Federated learning, each client/silo trains locally, Each client updates its own model, and The server averages all models.

### The problem with non-IID 
### 1. Client 1 only sees DDos Attacks, Client 2 sees brute force attacks, Client 3 mostly sees benign traffic

### The global minima becomes different from the local minima as for each client the goal becomes different.
### It is like figuring out the best study strategy. But, Student A only studeis maths, Student B only studies History and Student C only studies biology. 

### Each local /client update changes weights, Client 1 updates heavily DDos Features, whereas Client 2updates brute-force features.

---
### The solution to the problem: 
### We will be developint the global objective function, while respecting the privacy constraints of individual clients. 

### 

### We will be dividing the given frameowrk in two different stages: a Federated Deep Neural Network(DNN) pre training stage, and a federated novel attack fine tuning-stage., 

### In the pre training stage:
### Each silo trains its local DNN with local data, then uploads the weights to the central server,. The server then produces a global DNN model with weights via model weights averaging and broadcasts the global model back to each local silo.
