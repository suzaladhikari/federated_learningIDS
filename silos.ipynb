```python
import pandas as pd
```

```python
### Datasets we will be using 
monday_data = pd.read_csv('./datasets/Monday-WorkingHours.pcap_ISCX.csv')
print(f"The Monday dataset has {monday_data.shape[0]} number of rows, and {monday_data.shape[1]} number of columns")
print(monday_data[' Label'].value_counts())
```

```python
tuesday_data = pd.read_csv('./datasets/Tuesday-WorkingHours.pcap_ISCX.csv')
print(f"The Tuesday dataset has {tuesday_data.shape[0]} number of rows, and {tuesday_data.shape[1]} number of columns")
print(tuesday_data[' Label'].value_counts())
```

```python
wednesday_data = pd.read_csv('./datasets/Wednesday-workingHours.pcap_ISCX.csv')
print(f"The Wednesday dataset has {wednesday_data.shape[0]} number of rows, and {wednesday_data.shape[1]} number of columns")
print(wednesday_data[' Label'].value_counts())
```

```python
thursday_one = pd.read_csv('./datasets/Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv')
thursday_two = pd.read_csv('./datasets/Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv')
thursday_three = pd.read_csv('./datasets//Tuesday-WorkingHours.pcap_ISCX.csv')

print(thursday_one[' Label'].value_counts())
print(thursday_two[' Label'].value_counts())
print(thursday_three[' Label'].value_counts())

```

```python
### Combining all the three thursdays dataset 
thursday_dataset = pd.concat([thursday_one,thursday_two,thursday_three], ignore_index=True)
thursday_dataset.head(5)
print(f"Thursday dataset has {thursday_dataset.shape[0]} rows and {thursday_dataset.shape[1]} columns")
print(thursday_dataset[' Label'].value_counts())
```

```python
friday_one = pd.read_csv('./datasets/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
friday_two = pd.read_csv('./datasets/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv')
friday_three = pd.read_csv('./datasets/Friday-WorkingHours-Morning.pcap_ISCX.csv')

print(friday_one[' Label'].value_counts())
print(friday_two[' Label'].value_counts())
print(friday_three[' Label'].value_counts())
```

```python
friday_dataset = pd.concat([friday_one,friday_two,friday_three], ignore_index=True)
friday_dataset.head(5)
print(f"Thursday dataset has {friday_dataset.shape[0]} rows and {friday_dataset.shape[1]} columns")
print(friday_dataset[' Label'].value_counts())
```

### Combining all the dataset ! 

```python
data = pd.concat([monday_data,tuesday_data,wednesday_data,thursday_dataset,friday_dataset], ignore_index=True)
print(f"The final data has {data.shape[0]} rows and {data.shape[1]} columns in total")
print(data[' Label'].value_counts())
```

### We can see that there is severe class balance, where the highest tally is for Benign cases with 2.7 million samples whereas the lowest is among the attacks such heartbleed with about 11
